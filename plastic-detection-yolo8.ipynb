```python
from ultralytics import YOLO
import cv2
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
```

```python
# Loaв YOLOv8 model 
model = YOLO('yolov8n.pt')  
# Train model
data_path = os.path.abspath('data_yolo/data.yaml')
model.train(
    data=data_path,        
    epochs=50,                         
    imgsz=128,                         
    batch=-1                           
)
```

```python
# Load and preprocess the test image
image_path = "plastic-waste-1.jpg"  # Path to your test image
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Run inference
results = model.predict(source=image_path, save=False, show=False)

# Extract predictions
detections = results[0].boxes.xyxy.numpy()  # Bounding boxes (x1, y1, x2, y2)
scores = results[0].boxes.conf.numpy()     # Confidence scores
classes = results[0].boxes.cls.numpy()     # Class IDs
```

```python
import matplotlib.pyplot as plt
# Display results
for bbox, score, cls_id in zip(detections, scores, classes):
    x1, y1, x2, y2 = bbox
    label = f"{model.names[int(cls_id)]} {score:.2f}"
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
    cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Save the image with detections
output_path = "result_with_detections.jpg"  # Define output file path
# cv2.imwrite(output_path, image)  # Save the image
```

```python

# Display the image with detections
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
import random
# Display 10 random 128x128 images from the dataset
def display_random_images(dataset_path):
    # Get all image file names from the dataset folder
    image_files = [f for f in os.listdir(dataset_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
    
    if not image_files:
        print("No images found in the dataset folder. Please check the path.")
        return
    
    # Randomly select 10 images
    random_images = random.sample(image_files, min(10, len(image_files)))

    # Display the selected images
    plt.figure(figsize=(15, 15))
    for i, image_name in enumerate(random_images):
        image_path = os.path.join(dataset_path, image_name)
        image = cv2.imread(image_path)
        if image is None:
            print(f"Failed to load image: {image_path}")
            continue
        
        resized_image = cv2.resize(image, (128, 128))
        plt.subplot(2, 5, i + 1)
        plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
        plt.title(image_name)
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
display_random_images("data_yolo/dataset/images/train")
```
